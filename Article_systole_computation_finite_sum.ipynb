```python
def translation_length(M): #computes the translation length given a matrix
    trace = M.trace()
    l_1 = acosh(trace/2)
    l_1 = l_1.real()
    length = abs(2*l_1)
    l = float(length)
    
    return l
```

```python
import math #computes where the initial triple (0,i,inf) go when applying M
def mobiustransf(M):
    if M[1][1] == 0:
        #print("Parabolic element: common vertex at infinity")
        z_1 = math.inf
    else: 
        z_1 = M[0][1]/M[1][1]
    
    if M[1][0]*1J+M[1][1] == 0:     
        #print("Parabolic element: common vertex at infinity")
        z_2 = math.inf
    else:
        z_2 = (M[0][0]*I+M[0][1])/(M[1][0]*I+M[1][1])
       
    if M[1][0] == 0:  
        #print("Parabolic element: common vertex at infinity")
        z_3 = math.inf
    else:
        z_3 = M[0][0]/M[1][0]
        
    return z_1, z_2, z_3
```

```python
def center_radius(x1, y1, x2, y2, x3, y3): #function needed to compute the w-distance 
    A = x1 * (y2 - y3) - y1 * (x2 - x3) + x2 * y3 - x3 * y2
    B = (x1^2 + y1^2) * (y3 - y2) + (x2^2 + y2^2) * (y1 - y3) + (x3^2 + y3^2) * (y2 - y1)
    C = (x1^2 + y1^2) * (x2 - x3) + (x2^2 + y2^2) * (x3 - x1) + (x3^2 + y3^2) * (x1 - x2)
    D = (x1^2 + y1^2) * (x3*y2 - x2*y3) + (x2^2 + y2^2) * (x1*y3 - x3*y1) + (x3^2 + y3^2) * (x2*y1 - x1*y2)
    c_1 = -B/(A*2)
    c_2 = -C/(A*2)
    r = sqrt((B^2+C^2-4*A*D)/(4*A^2))
    return c_1, c_2, r
```

```python
def w_distance(w): #computes the w-distance of a matrix w
    import math
    z_1, z_2, z_3 = mobiustransf(w) #computes de Mobius transf of M: 0->z_1 , i->z_2, inf->z_3
    vertices = [0,I,math.inf]
      
    if ((z_1 in vertices) or (z_2 in vertices) or (z_3 in vertices)):
        hyp_dist = 0
        
    else: 
        c_1, c_2, r =  center_radius(real(z_1), imag(z_1), real(z_2), imag(z_2), real(z_3), imag(z_3)) #compute circle given the 3 pts
        R_sq = c_1^2 - r^2 #radius of semicircle corresp to the unique geodesic orthogonal to both planes
        x_1 = 0 #(x_1, x_2) is the intersection point btw the plane (0,i,inf) and the orthogonal geod
        x_2 = sqrt(R_sq)
        y_1 = (R_sq - r^2 + c_1^2)/(2*c_1)
        y_2 = sqrt(R_sq - y_1^2) #(x_1, x_2) is the intersection point btw the plane (z_1,z_2,z_3) and the orthogonal geod
        hyp_dist = acosh(1+((y_1 - x_1)^2 + (y_2 - x_2)^2)/(2*x_2*y_2)) #hyperbolic distance between points x and y
        
    return hyp_dist
```

```python
def next_parab(L): 
    for i in range(len(L)):
        L[i].insert(0, L[i][0])
    return L
```

```python
from collections.abc import Iterable

def flatten(xs):
    for x in xs:
        if isinstance(x, Iterable) and not isinstance(x, (str, bytes)):
            yield from flatten(x)
        else:
            yield x
```

```python
def list_traces(list):
    traces = []
    for i in range(len(list)):
        PM = identity_matrix(2)
        for j in range(len(list[i])):
            PM = PM*M[list[i][j]] #product of matrices M[j] with j running through each 3-tuple of indices (i) in list
        trace = PM.trace()
        traces.append(trace)
        
    return traces
```

```python
def word_backward_list(L, n):
    l = len(L)
    for k in range(l):    
        w_back = [0] * len(L[k])
        for i in [1..len(L[k])]:  
            index =  len(L[k])- i + 1
            im = Mod(index, n) #this is to include the case of change in first letter
            if L[k][len(L[k])-i] in [0,1,2]: #look at the orientation of the previous letter 
                if L[k][im] in [0,3,6]: #look at the letter we start with
                    w_back[len(L[k])-1-im]=0 #we put this new letter in the contrary position as onwards
                elif L[k][im] in [1,4,7]: 
                    w_back[len(L[k])-1-im]=1
                else:
                    w_back[len(L[k])-1-im]=2
            if L[k][len(L[k])-i] in [3,4,5]: 
                if L[k][im] in [0,3,6]:
                    w_back[len(L[k])-1-im]=3
                elif L[k][im] in [1,4,7]: 
                    w_back[len(L[k])-1-im]=4
                else:
                    w_back[len(L[k])-1-im]=5
            if L[k][len(L[k])-i] in [6,7,8]: 
                if L[k][im] in [0,3,6]:
                    w_back[len(L[k])-1-im]=6
                elif L[k][im] in [1,4,7]: 
                    w_back[len(L[k])-1-im]=7
                else:
                    w_back[len(L[k])-1-im]=8
        L.append(w_back)    
                
    return L
```

```python
def compute_words(w, i, L): #compute de equiv word after a change in orientation in the octahedron N-i
    w_new = [0] * len(w)
    for j in range(2):
        if (w[len(w)-1-i] == 2 or w[len(w)-1-i] == 5 or w[len(w)-1-i] == 8): #change of letter and orientation
            a_1 = w[len(w)-1-i]+1
            a = Mod(a_1, 9)
            w_new[len(w)-1-i] = a
        else: 
            a_1 = w[len(w)-1-i]+4
            a = Mod(a_1, 9)
            w_new[len(w)-1-i] = a
        
        if i == len(w)-1: #change of (only) orientation of previous letter
            b_1 = w[len(w)-1]+3
            b = Mod(b_1, 9)
            w_new[len(w)-1] = b
        else: 
            b_1 = w[len(w)-2-i]+3
            b = Mod(b_1, 9)
            w_new[len(w)-2-i] = b
            
        if not (i==len(w)-1 or i==len(w)-2): #non change of the other letters in the word
            for j in range(len(w)-i-2):
                w_new[j] = w[j]
        if not i==0:
            for j in [len(w)-i..len(w)-2]:
                w_new[j] = w[j]
            if not i==len(w)-1:
                w_new[len(w)-1] = w[len(w)-1]
       
        L.append(w_new)
        w = w_new
        w_new = [0] * len(w)
        
    return L
```

```python
def equiv_words_orient(w): #computes the list of equiv by all changes in orientation of all octahedra in a word w
    j=2
    L=[w]
    L = compute_words(w, len(w)-1, L)
    while j<len(w)+1:
        for i in range(len(L)):
            L = compute_words(L[i], len(w)-j, L)
        j = j+1
        
    return L
```

```python
def cycl_permute(L,k): #kth-cyclic permutation of a list
    k0 = k
    while k0 >= len(L):
        k0 = k0 - len(L)
                     
    Lnew = []
    for m in range(k0):
        Lnew += [L[len(L)-k0+m]]
    for m in range(len(L)-k0):
        Lnew += [L[m]]
   
    return Lnew

def are_cycl_perm(L1,L2): #check if two list are a cyclic permut of each other
    if len(L1) != len(L2):
        return False
    else:
        for k in range(len(L1)):
            if cycl_permute(L1,k) == L2:
                return True
    return False
```

```python
def cyclic_words(L): #Given a list of words L, computes all cyclic permutations of each word in L
    n=len(L)
    for i in range(n): #computes all cyclic permutations of each word in L
        for k in [1..len(L[i])-1]:
            L_cyclic = cycl_permute(L[i],k)
            L.append(L_cyclic) #it adds them in L
    return L
```

```python
#FUNCTION THAT COMPUTES THE LIST OF EQUIVALENT CLASSES OUT OF A LIST OF WORDS (using trace exclusion)
def list_equiv_classes_bytrace(L, trace_w):
    Cls = [] #list of equiv classes of words that we are gonna compute next
    tr =[] #list of traces of classes in Cls
    reduced_class = []
    Class0 = []
    Class_new = []


    tr.append(trace_w[0])
    Class0 = equiv_words_orient(L[0]) #Computes equiv words changing orientation of 1st element in L_index - returns a list with these words
    Class0 = word_backward_list(Class0, len(Class0[0])) #computes the word backwards of every element in the list Class0
    Class0 = cyclic_words(Class0) #compute all cyclic permutations of all words in Class0 - the equiv class is now completed
    [reduced_class.append(x) for x in Class0 if x not in reduced_class] #takes out the duplicates inside the class

    Cls.append(reduced_class.copy()) #add the first equiv class of words to the list Cls

    for i in [1..len(L)-1]:
        reduced_class.clear()
        new = 1
        negtrace = -trace_w[i]
        if trace_w[i] in tr:  #check if the trace of the next word is equal to the one of the previous words
            ind = len(tr) - 1 - tr[::-1].index(trace_w[i])
            if len(L[i]) == len(Cls[ind][0]):
                new = 0 #the word is not new
        elif negtrace in tr: #check if the -trace of the next word is equal to the one of the previous words
            ind = len(tr) - 1 - tr[::-1].index(trace_w[i])
            if len(L[i]) == len(Cls[ind][0]): #check if the length of the words of equal trace are equal 
                new = 0 #the word is not new
            
        if new == 1: #if it's new, do the same procedure as before to construct this new equiv class
            Class_new = equiv_words_orient(L[i]) 
            Class_new = word_backward_list(Class_new, len(Class_new[0]))
            Class_new = cyclic_words(Class_new)
            tr.append(trace_w[i]) 
        
            for x in Class_new: 
                if x not in reduced_class:
                    reduced_class.append(x)
  
            Cls.append(reduced_class.copy())
            print("Class done n=", i)    
        
    return Cls, tr
```

```python
#FUNCTION THAT COMPUTES THE LIST OF EQUIVALENT CLASSES OUT OF A LIST OF WORDS (by computing and comparing the class at each step)
def equiv_classes_in(L):
    reduced_class = []
    Cls = [] #list of equiv classes of words that we are gonna compute next
    Class0 = []
    Class_new = []

    Class0 = equiv_words_orient(L[0]) #Computes equiv words changing orientation of 1st element in L_index - returns a list with these words
    Class0 = word_backward_list(Class0, len(Class0[0])) #computes the word backwards of every element in the list Class0
    Class0 = cyclic_words(Class0) #compute all cyclic permutations of all words in Class0 - the equiv class is now completed
    [reduced_class.append(x) for x in Class0 if x not in reduced_class] #takes out the duplicates inside the class

    Cls.append(reduced_class.copy()) #add the first equiv class of words to the list Classes
    
    count = 0
    number = 0
    for i in [1..len(L)-1]:
        reduced_class.clear()
        new = 1
        for j in [0..count]:
            if L[i] in Cls[j]: #check if the next word in L_index is in the list of equiv words of the previous words
                new = 0 #the word is not new
                number = number + 1
                break
        if new == 1: #if it's new, do the same procedure as before to construct this new equiv class
            Class_new = equiv_words_orient(L[i]) 
            Class_new = word_backward_list(Class_new, len(Class_new[0]))
            Class_new = cyclic_words(Class_new) 
        
            for x in Class_new: 
                if x not in reduced_class:
                    reduced_class.append(x)
                
    
            count = count + 1
            Cls.append(reduced_class.copy())
            print("Class done n=", i)
        
    return Cls
```

```python
def translation_length_fromtrace(tr): #computes the translation length given a trace
    #trace = M.trace()
    l_1 = acosh(tr/2)
    l_1 = l_1.real()
    length = abs(2*l_1)
    l = length.n(digits=21)
    #l = float(length)

    return l
```

```python
#FUNCTION THAT COMPUTES THE SUM OF FIRSTS TERMS FOR SYSTOLE (for l<4.6)
def SUM_SYS(Classes, Traces, lengths_classes): 
    
    lengths = [] 
    Classes_of_length=[] #list of classes in Classes of length l_i
    trl=[] #list of lengths of all classes in list Classes
    pl=[] #list with all coef of the sum 
    Ord_Classes =[] #list of classes in list "Classes" ordered by translation length
    Ord_lengths_classes =[] #list of classes in list "Classes" ordered by translation length
    e = 1 #coefficient correspondent to "there is no word of length less than l_i"
    Card_classes_of_length = []
    
    for k in range(len(Traces)):
        l = translation_length_fromtrace(Traces[k]) #computes the list of lengths of the classes in list Classes
        trl.append(l)
    print("trl=", trl)

    Ord_Classes = [x for _, x in sorted(zip(trl, Classes))] #order the list Classes in terms of their translation length
    print("Ord_Classes=", Ord_Classes)

    Ord_lengths_classes = [x for _, x in sorted(zip(trl, lengths_classes))] #order the list lengths_classes in terms of their translation length
    print("Ord_lengths_classes=", Ord_lengths_classes)

    trl.sort() #now we order the list of lengths trl, so that it goes in relation with Ord_Classes
    print("sorted_trl=", trl)

    [lengths.append(x) for x in trl if x not in lengths] #takes out the duplicates inside the class -> list of lengths computed
    print("lengths=", lengths)
    
    c=0
    for i in [0..len(lengths)-1]:
        for j in [c..len(Ord_Classes)-1]: 
            if trl[j] == lengths[i]:
                Classes_of_length.append(Ord_Classes[j]) #we compute the list "Classes_of_length" 
                Card_classes_of_length.append(Ord_lengths_classes[j])
                print("Classes_of_length=", Classes_of_length)
            else:
                break #since the classes are ordered, once it is not equal means it's bigger, so we stop comparing
        c=j

        e_new=1
        for t in range(len(Classes_of_length)):
            r = len(Classes_of_length[t])
            e_new = e_new* exp(-Card_classes_of_length[t]/(3^r*2*r)) #compute the exponent for words of length l_i and word_length r

        coef = e*(1-e_new)*lengths[i] #1st coefficient p_1*l_1 of the sum
        print("coef=", coef)
        pl.append(coef)
        #print("pl=", pl)
        e = e*e_new #the coefficient e passes to be the e_new

        Classes_of_length.clear() #we clear the list to fill it with the next length
        Card_classes_of_length.clear()

    Sum = sum(pl)
    S_pl = Sum.n(digits=21) 

    return Sum, S_pl
```
